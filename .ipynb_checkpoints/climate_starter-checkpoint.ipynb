# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()

# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
M = Base.classes.measurement
S = Base.classes.station
```

```python

# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Climate Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results

d = session.query(M.date).order_by(M.date.desc()).first()
d
```

```python
# Calculate the date 1 year ago from the last data point in the database

y = dt.date(2017,8,23) - dt.timedelta(days=365)
y

```

```python
# Perform a query to retrieve the data and precipitation scores

# precipitation score (pcrp)

# I calculated these from 2016-08-23 using M.date >=y. I interpreted the request as such. 
pcrp_score = session.query(M.date, M.prcp).filter(M.date >= y).\
    order_by(M.date).all()
pcrp_score
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column
# Sort the dataframe by date
pcrp_score1 = pd.DataFrame(pcrp_score, columns=["Date","Precipitation (Inches)"])
pcrp_score1.set_index("Date", inplace=True,)
pcrp_score1.head()
```

```python
# Use Pandas Plotting with Matplotlib to plot the data

pcrp_score1.plot(figsize=(20,10))
plt.legend(loc='upper right')
plt.xlabel("Date")
plt.ylabel("Inches")
plt.savefig("Images/rain.png")
plt.show()
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(M.tobs), func.avg(M.tobs), func.max(M.tobs)).\
        filter(M.date >= start_date).filter(M.date <= end_date).all()

# function usage example

print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.

start = '2016-06-01' 
end = '2016-06-14' 
trip = calc_temps(start, end)
trip
```

```python
# Plot the results from your previous query as a bar chart. 
# Use "Trip Avg Temp" as your Title
# Use the average temperature for the y value
# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)

trip1 = pd.DataFrame(trip, columns=["TMIN", "TAVG", "TMAX"])
trip1.plot.bar( y="TAVG", yerr=(trip1["TMAX"] - trip1["TMIN"]), 
                    title="Trip Avg Temp", color="coral", 
                    alpha=0.5, figsize=(5,7), legend=None)
plt.xticks([])
plt.ylabel("Temp (F)")
plt.tight_layout()
plt.savefig("Images/trip_avg_temp.png")
plt.show()

```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(M.tobs), func.avg(M.tobs), func.max(M.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", M.date) == date).all()
    
daily_normals("01-01")
```

```python

# calculate the daily normals for your trip
# push each tuple of calculations into a list called `normals`

trip1 =['06-01','06-02','06-03','06-04','06-05','06-06','06-07']

normals=[]
    
def daily_normals(date):
    sel = [func.min(M.tobs), func.round(func.avg(M.tobs),2), func.max(M.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", M.date) == date).all()

for day in trip1:
    normals.append(daily_normals(day)[0])

normals
```

```python
# Calculate the normal for each trip date

normals1=pd.DataFrame(normals,columns=['TMIN','TAVG','TMAX'], index=trip1)

normals1.index.name='date'
normals1
```

```python
plt.rcParams['figure.figsize']=(11,7)

normals1.plot.area(linewidth=4,stacked=False,rot=0,alpha=0.5);

plt.ylim(-10,100)
plt.yticks(size=14)
plt.xticks(fontsize=14)
plt.xlabel('Date (mm-dd)',fontsize=14,color='black')
plt.ylabel('Temperature (F)',fontsize=16,color='black')
plt.title('Daily Temp Normals',fontsize=25,pad=40)
plt.gca().legend(loc='center', bbox_to_anchor=(1, 1),shadow=True,borderpad=1);
```
